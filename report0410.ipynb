```python
#2-1)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


x = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['x']).values.flatten()
y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['y']).values.flatten()


plt.scatter(x, y)
plt.show()


w1 = 0.0
w0 = 0.0
learning_rate = 0.001
epochs = 500

n = len(x)


for epoch in range(epochs):
    y_pred = w1 * x + w0
    dw1 = (1/n) * sum(x * (y_pred - y))
    dw0 = (1/n) * sum(y_pred - y)
    w1 = w1 - learning_rate * dw1
    w0 = w0 - learning_rate * dw0
print(':')
print(f"w1 = {w1:.4f}")
print(f"w0 = {w0:.4f}")



plt.figure(figsize=(10, 6))
plt.title('Linear Regression Result')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y, label='Data', color='blue')

x_line = np.linspace(min(x), max(x), 300)
y_line = w1 * x_line + w0
plt.plot(x_line, y_line, color='red', label='Fitted Line', linewidth=2)

plt.legend()
plt.grid(True)
plt.show()
```

```python
#2-2)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


x = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['x']).values.flatten()
y = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data0405.csv', usecols=['y']).values.flatten()


def polynomial_regression(x, y, degree, learning_rate=0.001, epochs=500):
    x1 = np.vstack([x**d for d in range(degree, -1, -1)]).T
    w = np.zeros(degree + 1)
    n = float(len(x))

    for i in range(epochs):
        y_pred = x1.dot(w)
        error = y - y_pred
        gradient = -(2/n) * x1.T.dot(error)
        w = w - learning_rate * gradient
        if i % 100 == 0:
            mse = np.mean(error**2)
    return w


def predict(x, w):
    degree = len(w) - 1
    return sum(w[i] * x**(degree - i) for i in range(len(w)))


w2 = polynomial_regression(x, y, degree=2)
w4 = polynomial_regression(x, y, degree=4)


xp = np.linspace(min(x), max(x), 300)
y2 = predict(xp, w2)
y4 = predict(xp, w4)


plt.figure(figsize=(10, 8))


plt.subplot(2, 1, 1)
plt.scatter(x, y, color='blue', label='Data', alpha=0.6)
plt.plot(xp, y2, color='red', linewidth=2, label='2nd_Nonlinear')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)


plt.subplot(2, 1, 2)
plt.scatter(x, y, color='blue', label='Data', alpha=0.6)
plt.plot(xp, y4, color='green', linewidth=2, label='4th_Nonlinear')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

print("2nd_param:")
for i, c in enumerate(w2):
    print(f"a{i} = {c:.4f}")

print("\n4th_param:")
for i, c in enumerate(w4):
    print(f"a{i} = {c:.4f}")


plt.tight_layout()
plt.show()
```
